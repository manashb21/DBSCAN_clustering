```python
import numpy as np
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt 
import pandas as pd

X,_ = make_blobs(n_samples = 500, centers = 3,  n_features = 2, random_state = 20)
```

```python
X
```

```python
df = pd.DataFrame(dict(x=X[:,0], y=X[:,1]))
fig,ax = plt.subplots(figsize=(8,8))
df.plot(ax = ax, kind ='scatter', x = 'x', y = 'y')
```

```python
from sklearn.cluster import DBSCAN
import random
from sklearn import metrics

cluster = DBSCAN(eps = 1, min_samples = 5).fit(X)
labels = cluster.labels_
unique_labels = np.unique(labels)
```

```python

color = ['tomato','skyblue','purple','aquamarine','teal']
df
def show_clusters(X,labels):
    df = pd.DataFrame(dict(x=X[:,0],y=X[:,1], labels=labels))
    fig, ax = plt.subplots(figsize =(8,8))
    grouped = df.groupby('labels')
    for key, group in grouped:
        group.plot(ax=ax, kind ='scatter', x='x', y='y', label = key, color = color[key])
    plt.show()
```

```python
show_clusters(X,labels)
print('Silhouette Score: ',metrics.silhouette_score(X,labels))
```
